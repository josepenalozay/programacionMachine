**Series**

Un array unidimensional de datos (de cualquier tipo numpy), con valores indexados. 
Se puede crear pasando una lista o dict de Python, una matriz de numpy, un archivo csv


```python
pd.Series([1,3,np.nan]) 
# Default integers are integers
```

```python
pd.Series([1,3,5], index=['a','b','c'])
```

```python
pd.Series({'a' : 1, 'b': 2, 'c': 3 }) 
# when given a dict, the keys will be used for the index
```

```python
dicci = pd.Series({'a' : 1, 'b': 2, 'c': 3 }, index = ['b', 'c', 'd']) 
# this will try to match labels with keys
```

```python
dicci
```

```python
#print(series_from_list[0]) 
print(dicci['b'])
```

```python
a = pd.Series({'John' : 1000, 'Mary': 2000, 'Andre': 3000 })
b = pd.Series({'John' : 100, 'Andre': 200, 'Cecilia': 300 })
a + b
```

```python
series_a = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
series_b = pd.Series([4, 5, 6], index=['b', 'c', 'd'])
sum_series = series_a + series_b 
print(sum_series)
```

**DataFrame**

Un DataFrame es una estructura de datos tabular con un índice de fila y otro de columna. 
Se puede crear pasando un dict de matrices, un archivo csv,...).


```python
data = {'state': ['Ohio', 'Ohio', 'Nevada', 'Nevada'], 
        'year': [2000, 2001, 2001, 2002],
        'pop': [1.5, 1.7, 2.4, 2.9]}
pd.DataFrame(data)
pd.DataFrame(data, columns=['year', 'state', 'pop', 'color']) 
# Will match indices
```

```python
i = 2
df.iloc[0]  # First row of a data frame
df.iloc[i]  #(i+1)th row 
df.iloc[-1] # Last row 
```

```python
df.iloc[:, 0]  # First column
df.iloc[:, -1] # Last column 
```

```python
df.iloc[0:7]       #First 7 rows 
df.iloc[:, 0:2]    #First 2 columns
df.iloc[1:3, 0:2]  #Second through third rows and first 2 columns
df.iloc[[0,5], [1,3]]  #1st and 6th rows and 2nd and 4th columns
```
