# Librerias

```python
import numpy as np
from  cv2 import imread
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
def howis(img):
  print('size = ',img.shape)
  print('max  = ',np.max(img))
  print('min  = ',np.min(img))
```

# Lectura de imagen a color

```python
!wget https://www.dropbox.com/s/jetlnfihles3g7k/flowers.jpg
```

```python
Icv = imread('flowers.jpg')      # lectura en formato opencv (BGR)
cv2_imshow(Icv)
howis(Icv)
```

```python
Icv
```

```python
plt.imshow(Icv)
plt.show()
```

```python
I = Icv[:,:,[2,1,0]]             # conversion a formato estándar (RGB)
plt.imshow(I)
plt.show()
```

# Canales de color

```python
R = I[:,:,0]
G = I[:,:,1]
B = I[:,:,2]
RGB = np.concatenate((R,G,B),axis=1)
cv2_imshow(RGB)
```

# Conversion a blanco y negro (tonos de gris)

```python
Rd = R.astype(float)
Gd = G.astype(float)
Bd = B.astype(float)
#k  = (1/3,1/3,1/3)
k  = (0.1,0.4,0.5)
Zd = k[0]*Rd+k[1]*Gd+k[2]*Bd
Z  = Zd.astype(int)
howis(Z)
cv2_imshow(Z)
```

# Histograma

```python
def imhist(X):
  (N,M) = X.shape
  n = 256
  h = np.zeros((256,))
  for i in range(N):
    for j in range(M):
      x = X[i,j]
      h[x] = h[x]+1
  plt.plot(range(n),h[0:n])
  plt.show()
```

```python
imhist(Z)
```
